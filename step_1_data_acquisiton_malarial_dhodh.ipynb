#### Import all necessay modules for collectinting  dataset from ChEMBL

```python
import pandas as pd
import numpy as np
from chembl_webresource_client.new_client import new_client
from rdkit.Chem import PandasTools
from zipfile import ZipFile
from tempfile import TemporaryDirectory
from tqdm.auto import tqdm
```

### Searching for our target protein Malaria Plasmodium falciparum dihydroorotate dehydrogenase (DHODH)


```python
target_protein = new_client.target
target_query = target_protein.search('Dihydroorotate dehydrogenase').filter(target_type='SINGLE PROTEIN').filter(organism='Plasmodium falciparum')
data = pd.DataFrame.from_dict(target_query)
data
```

#### The first one (0th entry) is our target protein that is Plasmodium falciparum Dihydroorotate dehydrogenase (PfDHODH)

```python
our_target = data.iloc[0]['target_chembl_id']
our_target
```

##### Getting compounds that inhibit PfDHODH target

```python
target_activity = new_client.activity
target_query = target_activity.filter(target_chembl_id=our_target).filter(standard_type='IC50')
```

```python
df = pd.DataFrame.from_dict(target_query)
```

```python
df.to_csv('pf_DHODH_01_bioactivity_data_raw.csv', index=False)
```

#### Classification of compound biological activity based on IC50 values

The bioactivity data is measured in IC50 units. Compounds with values below 1000 nM will be classified as active, those above 10,000 nM will be classified as inactive, and those falling between 1,000 and 10,000 nM will be classified as intermediate

```python
df4 = pd.read_csv('pf_DHODH_02_bioactivity_data_preprocessed.csv')
df4
```

```python
bioactivity_threshold = []
for i in df4.standard_value:
  if float(i) >= 10000:
    bioactivity_threshold.append("inactive")
  elif float(i) <= 1000:
    bioactivity_threshold.append("active")
  else:
    bioactivity_threshold.append("intermediate")
     
```

```python
bioactivity_class = pd.Series(bioactivity_threshold, name='class')
df5 = pd.concat([df4, bioactivity_class], axis=1)
df5
```
